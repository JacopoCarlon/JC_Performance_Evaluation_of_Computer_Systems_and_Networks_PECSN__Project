```python
maximum_bins_number = 10

# general_array[dst_srv][dst_tso].append(float(tempi_risposta[cont]))

save = False

plot_dir = f'../images/plots' 
for srvn in range(0, len(server_names)):
    for tso_this_iter in range(0, len(target_tso), 1) :
        plt.rcParams["figure.figsize"] = (25, 10)

        massimo_jResTime = float( max(general_array[srvn][tso_this_iter]) )
        
        counts, bins = np.histogram( general_array[srvn][tso_this_iter], bins = maximum_bins_number )

        counts = counts / len( general_array[srvn][tso_this_iter] ) # normalize count
        #   print("counts : ", counts)
        #   print("bins : ", bins)


        bin_width = bins[1]-bins[0]
        hbw = bin_width / 2
        #   bins = bins - hbw

        plt.hist(bins[:-1], bins, weights=counts, edgecolor='black')
        ax = plt.gca()
        ax.yaxis.get_major_formatter().set_scientific(False)
        ax.yaxis.get_major_formatter().set_useOffset(False)
        ax.xaxis.get_major_formatter().set_useOffset(False)
        ax.xaxis.get_major_formatter().set_scientific(False)

        extras = ["fast", "slow"]
        #   plt.hist(counts)
        name = f'jResponseTime_ExponentialTSO_ConstantTN{target_tso[tso_this_iter]}_P{probability}_Server{srvn}({extras[srvn]})'
        
        plt.title(f"{name}")
               
        plt.ylabel("Probability")
        plt.xlabel("Response Time (seconds)")
        

        # print mean !
        target_asNpArray = np.array(general_array[srvn][tso_this_iter])
        print(target_asNpArray)
        target_mean = target_asNpArray.mean()
        print(target_mean)
        plt.axvline( target_mean , color='k', linestyle='dashed', linewidth=1 )
        min_ylim, max_ylim = plt.ylim()
        
        text_mean_location = target_mean + hbw
        plt.text( text_mean_location, max_ylim*0.97, 'Mean: {:.2f} s'.format(target_mean))


        
        endingPNG = ".png"
        if detailed == True:
            endingPNG = "_isDetailed.png"

        if save == True:
            plt.savefig(f'{plot_dir}/jResTime_plot{distribution}P{probability}_{target_tso[tso_this_iter]}_onlyIterZero_Server{srvn}{endingPNG}')
        
        plt.show()

        #   if tso_this_iter > 1:
        #       break
    
```
